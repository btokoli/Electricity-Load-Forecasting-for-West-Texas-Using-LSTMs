```python
# 1) Define the region mapping for your lat/lon
region_map = {
    "32.01_-102.06": "WEST",
    "31.53_-97.14":  "NORTH_C",
    "31.77_-106.50": "FAR_WEST",
    "32.77_-96.78":  "NORTH"
}

# 2) Create a dictionary listing all four files for each region
#    (Make sure the filenames match exactly what you see in !ls)
region_files = {
    "WEST": [
        "560977_32.01_-102.06_2012.csv",
        "560977_32.01_-102.06_2013.csv",
        "560977_32.01_-102.06_2014.csv",
        "560977_32.01_-102.06_2015.csv"
    ],
    "NORTH_C": [
        "694093_31.53_-97.14_2012.csv",
        "694093_31.53_-97.14_2013.csv",
        "694093_31.53_-97.14_2014.csv",
        "694093_31.53_-97.14_2015.csv"
    ],
    "FAR_WEST": [
        "447515_31.77_-106.50_2012.csv",
        "447515_31.77_-106.50_2013.csv",
        "447515_31.77_-106.50_2014.csv",
        "447515_31.77_-106.50_2015.csv"
    ],
    "NORTH": [
        "702800_32.77_-96.78_2012.csv",
        "702800_32.77_-96.78_2013.csv",
        "702800_32.77_-96.78_2014.csv",
        "702800_32.77_-96.78_2015.csv"
    ]
}

# 3) Combine files by region
region_dfs = {}  # will store each combined DataFrame keyed by region name

for region, file_list in region_files.items():
    df_list = []
    for filename in file_list:
        # Read each CSV into a temporary DataFrame
        temp_df = pd.read_csv(filename)
        df_list.append(temp_df)

    # Concatenate the 4 DataFrames (one per year) into one
    combined_df = pd.concat(df_list, ignore_index=True)

    # Add a "region" column for clarity
    combined_df["region"] = region

    # Store in a dictionary
    region_dfs[region] = combined_df

    print(f"Combined {len(file_list)} files for {region}. Shape: {combined_df.shape}")

# Each region is now asingular data frame
west_df     = region_dfs["WEST"]
north_c_df  = region_dfs["NORTH_C"]
far_west_df = region_dfs["FAR_WEST"]
north_df    = region_dfs["NORTH"]

# Save each combined DataFrame as CSV
for region, df in region_dfs.items():
    out_name = f"{region}_combined.csv"
    df.to_csv(out_name, index=False)
    print(f"Saved {out_name}")
```

```python
import pandas as pd
from IPython.display import display

# The columns you want in each final DataFrame
cols_to_keep = [
    "Year", "Month", "Day", "Hour",
    "Temperature", "Dew Point", "GHI",
    "Wind Speed", "Relative Humidity", "region"
]

# A dictionary listing each region's CSV file + region name
region_files = {
    "NORTH_C": {
        "filename": "NORTH_C_combined.csv",
        "region_name": "NORTH_C"
    },
    "FAR_WEST": {
        "filename": "FAR_WEST_combined.csv",
        "region_name": "FAR_WEST"
    },
    "NORTH": {
        "filename": "NORTH_combined.csv",
        "region_name": "NORTH"
    },
    "WEST": {
        "filename": "WEST_combined.csv",
        "region_name": "WEST"
    }
}

# Dictionary to store the final DataFrames
region_dfs = {}

for region_key, info in region_files.items():
    file_path = info["filename"]
    region_name = info["region_name"]

    # Read the CSV, skipping the first 2 lines of metadata
    df = pd.read_csv(
        file_path,
        skiprows=2,   # skip the "NSRDB" row + the partial metadata row
        header=0,     # use the third row as column headers
        low_memory=False
    )

    # Ensure a 'region' column is present
    if "region" not in df.columns:
        df["region"] = region_name

    # Subset columns (make sure they actually exist in df.columns)
    # If any are missing, you'll get a KeyError.
    # Adjust if 'Dew Point' or others are spelled differently.
    df = df[cols_to_keep]

    # Store it
    region_dfs[region_key] = df

    # Show results
    print(f"\n=== {region_key} DataFrame ===")
    print("Shape:", df.shape)
    display(df.head(5))
```
